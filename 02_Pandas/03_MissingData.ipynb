```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'A':[1,2,np.nan,4],
                   'B':[5,np.nan,np.nan,8],
                   'C':[10,20,30,40]})
```

```python
df
```

```python
df.dropna(axis=0)
```

```python
df.dropna(axis=1)
```

```python
df.dropna(thresh=2)
```

```python
df['A'].fillna(value=0)
```

```python
df['A'].fillna(value=df['A'].mean())
```

```python
df.fillna(df.mean())
```
